```python
import cv2
import os

path = 'C:\\Users\\SASG\\Downloads\\potholes_datasets\\roaddamage21\\'
out_path = 'C:\\Users\\SASG\Downloads\\img_to_video\\'
out_video_name = 'roaddamage21frames2.mp4'
out_video_full_path = out_path+out_video_name

pre_imgs = os.listdir(path)
# print(pre_imgs)
img = []

for i in pre_imgs:
    i = path+i
    # print(i)
    img.append(i)

# print(img)

cv2_fourcc = cv2.VideoWriter_fourcc(*'mp4v')

frame = cv2.imread(img[0])
size = list(frame.shape)
del size[2]
size.reverse()
print(size)

video = cv2.VideoWriter(out_video_full_path, cv2_fourcc, 2, size) #output video name, fourcc, fps, size

for i in range(len(img)): 
    video.write(cv2.imread(img[i]))
    print('frame ', i+1, ' of ', len(img))

video.release()
print('outputed video to ', out_path)
```

```python
image_folder = 'C:\\Users\\SASG\\Downloads\\img_sequence'
output_size = (450, 300)  # Set the desired size for all images
```

```python
from PIL import Image
# Resize images and save them temporarily
resized_image_folder = 'C:\\Users\\SASG\Downloads\\img_to_video'
os.makedirs(resized_image_folder, exist_ok=True)

for img_name in os.listdir(image_folder):
    if img_name.endswith(".png"):
        img_path = os.path.join(image_folder, img_name)
        img = Image.open(img_path).convert('RGB')  # Convert to RGB mode
        img_resized = img.resize(output_size, Image.LANCZOS)  # Use LANCZOS as ANTIALIAS is deprecated
        img_resized.save(os.path.join(resized_image_folder, img_name))
```

```python
import tensorflow as tf
import os

def _parse_function(proto):
    # Define your features according to the structure of your TFRecord
    keys_to_features = {
        'image/encoded': tf.io.FixedLenFeature([], tf.string),  # For encoded images like JPEG/PNG
        # Add other features if necessary
    }

    # Load one example
    parsed_features = tf.io.parse_single_example(proto, keys_to_features)

    # Extract the image
    image = tf.io.decode_image(parsed_features['image/encoded'])
    
    return image

def extract_images_from_tfrecord(tfrecord_path, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Create a dataset from the TFRecord file
    dataset = tf.data.TFRecordDataset([tfrecord_path])

    # Map the parsing function to each example
    dataset = dataset.map(_parse_function)

    # Iterate through the dataset and save images
    for i, image in enumerate(dataset):
        # Convert the tensor to a numpy array
        img_array = image.numpy()
        
        # Save the image
        img_name = os.path.join(output_dir, f'image_{i+1}.png')
        tf.keras.preprocessing.image.save_img(img_name, img_array)
        print(f'Saved {img_name}')

# Example usage
tfrecord_path = 'C:\\Users\\SASG\\Downloads\\potholes_datasets\\roaddamage2\\train\\road-damage.tfrecord'
output_dir = 'C:\\Users\\SASG\\Downloads\\potholes_datasets'

extract_images_from_tfrecord(tfrecord_path, output_dir)
```

# Video to image

```python
import cv2
import os

# Path to the video file
video_path = 'D:\Pothole_Segmentation_YOLOv8\\sample_video.mp4'

# Directory to save the extracted images
output_dir = 'D:\\New folder\\New folder(sample_video)'
os.makedirs(output_dir, exist_ok=True)

# Open the video file
cap = cv2.VideoCapture(video_path)

# Initialize frame counter
frame_number = 0

# Loop through all frames in the video
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Save each frame as an image
    frame_filename = os.path.join(output_dir, f'frame_{frame_number:05d}.jpg')
    cv2.imwrite(frame_filename, frame)

    frame_number += 1

# Release the video capture object
cap.release()

print(f"Extracted {frame_number} frames from the video.")
```
